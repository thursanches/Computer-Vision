# Blending and Pasting Images

For some computer vision systems, we'll want to be able to post our own image on top of an already existing image or video. We may also want to blend images, maybe we want to have a "highlight" effect instead of just a solid box or empty rectangle.

Let's explore what is commonly known as **Arithmetic Image Operations** with OpenCV. These are referred to as Arithmetic Operations because OpenCV is simply performing some common math with the pixels for the final effect. We'll see a bit of this in our code.

## Blending Images

Blending Images is actually quite simple, let's look at a simple example.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Two images
imgdogpack = cv2.imread('dog_backpack.png')
imgwater = cv2.imread('watermark_no_copy.png')
```

```python
imgdogpack.shape
```

```python
imgwater.shape
```

```python
plt.imshow(imgdogpack)
```

Zuada de novo!!!

```python
imgdogpack = cv2.cvtColor(imgdogpack, cv2.COLOR_BGR2RGB)
imgwater = cv2.cvtColor(imgwater, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imgdogpack)
```

```python
plt.imshow(imgwater)
```

### Resizing the Images

```python
imgdogpack =cv2.resize(imgdogpack,(1200,1200))
imgwater =cv2.resize(imgwater,(1200,1200))
```

Let's practice resizing the image, since the DO NOT COPY image is actually quite large 1200 by 1200, and our puppy in backpack image is 1400 by 1000

```python
plt.imshow(imgdogpack)
```

```python
plt.imshow(imgwater)
```

### Blending the Image

We will blend the values together with the formula:

$$  img1 * \alpha  + img2 * \beta  + \gamma $$

```python
imgdogpack.shape
```

```python
imgwater.shape
```

```python
blended = cv2.addWeighted(src1=imgdogpack,alpha=0.7,src2=imgwater,
                          beta=0.3,gamma=0)
```

```python
plt.imshow(blended)
```

## Overlaying Images of Different Sizes

We can use this quick trick to quickly overlap different sized images, by simply reassigning the larger image's values to match the smaller image.

```python
# Load two images
imgdp = cv2.imread('dog_backpack.png')
imgwt = cv2.imread('watermark_no_copy.png')
imgwt =cv2.resize(imgwt,(600,600))

# fixing the image para nao ficar zuada!

imgdp = cv2.cvtColor(imgdp, cv2.COLOR_BGR2RGB)
imgwt = cv2.cvtColor(imgwt, cv2.COLOR_BGR2RGB)

large_img = imgdp
small_img = imgwt
```

```python
x_offset=0
y_offset=0
```

```python
large_img[y_offset:y_offset+small_img.shape[0],
          x_offset:x_offset+small_img.shape[1]] = small_img

# Explain the logic of that sintax
```

```python
plt.imshow(large_img)
```
